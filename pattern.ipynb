```python
n = int(input("Enter size: "))

for i in range(n):
    print("* " * n)
```

```python
n = int(input("Enter size: "))

for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
n = int(input("Enter number of rows: "))

num = 1
for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
n = int(input("Enter number of rows: "))

for i in range(n):
    ch = 65
    for j in range(i + 1):
        print(chr(ch), end=" ")
        ch += 1
    print()
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("* " * i)

for i in range(n - 1, 0, -1):
    print(" " * (n - i), end="")
    print("* " * i)
```

```python
n = int(input("Enter rows: "))

for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
n = 5

for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
